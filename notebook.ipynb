```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

from astropy.io import ascii

%matplotlib notebook
%config InlineBackend.figure_format = 'retina'
```

```python
sns.set_context('paper')
sns.set_style('dark')

t = ascii.read("./Spectra/transmission_CH4.dat", header_start=1)

wav, depth = t['(meters)'], t['(percent)']

plt.semilogx(wav*1E6, depth)
plt.show()
```
